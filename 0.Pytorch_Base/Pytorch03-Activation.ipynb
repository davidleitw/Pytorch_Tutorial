```python
import torch
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
# Build some data 
data = torch.linspace(-5,5,200) # -5 ~ +5 separate 200 part
npdata = data.numpy()           # Build np.array 
print(data)
print(data.size())
```

```python
y_relu = F.relu(data).numpy()

# Draw the graph for relu
plt.plot(npdata,y_relu,c='green',label='Relu')
plt.xlabel('x')
plt.ylabel('y')
plt.title(' Relu Function ')
plt.legend(loc='best')
plt.show()
```

```python
y_tanh = F.tanh(data).numpy()

# Draw the graph for tanh
plt.plot(npdata,y_tanh,c='green',label='Tanh')
plt.xlabel('x')
plt.ylabel('y')
plt.title(' Tanh Function ')
plt.legend(loc='best')
plt.show()
```

```python
y_sigmoid = F.sigmoid(data).numpy()

# Draw the graph for sigmoid
plt.plot(npdata,y_sigmoid,c='green',label='Sigmoid')
plt.xlabel('x')
plt.ylabel('y')
plt.title(' Sigmoid Function ')
plt.legend(loc='best')
plt.show()
```

```python
y_softplus = F.softplus(data).numpy()

# Draw the graph for softplus
plt.plot(npdata,y_softplus,c='green',label='Softplus')
plt.xlabel('x')
plt.ylabel('y')
plt.title(' Softplus Function ')
plt.legend(loc='best')
plt.show()
```

```python
# Draw the graph for relu

plt.subplot(221)
plt.plot(npdata,y_relu,c='green',label='Relu')
plt.xlabel('x')
plt.ylabel('y')
# plt.title(' Relu Function ')
plt.legend(loc='best')

# Draw the graph for tanh

plt.subplot(222)
plt.plot(npdata,y_tanh,c='green',label='Tanh')
plt.xlabel('x')
plt.ylabel('y')
# plt.title(' Tanh Function ')
plt.legend(loc='best')

# Draw the graph for softplus

plt.subplot(223)
plt.plot(npdata,y_softplus,c='green',label='Softplus')
plt.xlabel('x')
plt.ylabel('y')
# plt.title(' Softplus Function ')
plt.legend(loc='best')

# Draw the graph for sigmoid

plt.subplot(224)
plt.plot(npdata,y_sigmoid,c='green',label='Sigmoid')
plt.xlabel('x')
plt.ylabel('y')
# plt.title(' Sigmoid Function ')
plt.legend(loc='best')

plt.show()
```
